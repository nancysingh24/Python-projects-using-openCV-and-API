```python
#importing beautifulsoup as bs4 for web scrapping 
import bs4
import requests
#res variable stores the url content
res=requests.get("https://learncodeonline.in/")
#making object soup for storing beautifulsoup data in the text form 
soup=bs4.BeautifulSoup(res.text,"lxml")
#for loop is used for getting all links from the website
for link in soup.findAll("a",href=True):
    print(link["href"])
    
```

```python
from bs4 import BeautifulSoup
import requests
#making beautifulsoup object to get the content from the website
data=BeautifulSoup(requests.get("https://www.pexels.com/search/chocolate%20cakes/").content)
#mob variable stores all the images available in the website
mob=data.find_all("img")
#len() to find the total no of images
len(mob)
```

```python
#src is the empty list
src=[]
#appending  all the src tags in the src list
for i in range(len(mob)):
    src.append(mob[i]["src"])
```

```python
#to download image through web scrapping using beautifulsoup
import io
import requests
#PIL is used for manipulation of images
from PIL import Image
u="https://cdn.pixabay.com/photo/2015/02/24/15/41/dog-647528__340.jpg"
raw_data=requests.get(u,stream=True).content
array=io.BytesIO(raw_data)
#converting to array and rgb format and saving it as well
Image.open(array).convert("RGB").save("dog.jpg")
#array.find_all("img")[0]["src"]
Image.open(array).convert("RGB")
```

```python
from bs4 import BeautifulSoup
import requests

url = 'https://en.wikipedia.org/wiki/Transhumanism'
# get contents from url
content = requests.get(url).content
# get soup
soup = BeautifulSoup(content,'lxml') # choose lxml parser
# find the tag : <img ... >
image_tags = soup.findAll('img')
# print out image urls
for image_tag in image_tags:
    print(image_tag.get('src'))
```
