```python
import cv2 as cv
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import requests
import io
```

```python
image=cv.imread("fami.jpg")
plt.imshow(image)
```

```python
img_rgb=cv.cvtColor(image,cv.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
image_array=np.array(image)
```

```python
gray=cv.cvtColor(image_array,cv.COLOR_RGB2GRAY)
plt.imshow(gray,cmap="gray")
```

```python
detector=cv.CascadeClassifier("haar_frontal.xml")
```

```python
faces=detector.detectMultiScale(gray,1.3,5)#1.3-scale factor,5 is the distance btw 2 faces
```

```python
faces[0]
```

```python
x,y,w,h=faces[0]
```

```python
cv.rectangle(image_array,(x,y),(x+w,y+h),(255,255,255),2)
plt.imshow(image_array)
```

```python
faces[1]
```

```python
x,y,w,h=faces[1]
```

```python
cv.rectangle(image_array,(x,y),(x+w,y+h),(255,255,255),2)
plt.imshow(image_array)
```
