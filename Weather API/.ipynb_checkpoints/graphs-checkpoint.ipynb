```python
n1=np.arange(0,4*np.pi,0.1)
```

```python
n2=np.sin(n1)
plt.plot(n1,n2)
```

```python
def uni_step(x):
    if x<0:
        return 0
    else:
        return 1
```

```python
x=np.arange(-10,10)
```

```python
y=list(map(uni_step,x))
```

```python
y
```

```python
plt.plot(x,y)
```
