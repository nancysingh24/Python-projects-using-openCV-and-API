```python
import requests
import json
import numpy as np
import matplotlib.pyplot as plt
```

```python
caps=['Hyderabad',
 'Patna',
 'Agartala',
 'Kohima',
 'Chandigarh',
 'Ranchi',
 'Thiruvananthapuram',
 'Mumbai',
 'Bhubaneswar',
 'Chandigarh',
 'Panaji',
 'Gandhinagar',
 'Dispur',
 'Amaravati',
 'Dehradun',
 'Shimla',
 'Bhopal',
 'Jaipur',
 'Aizawl',
 'Srinagar',
 'Kolkata',
 'Raipur',
 'Chennai',
 'Lucknow',
 'Gangtok',
 'Itanagar',
 'Bangalore',
 'Shillong',
 'Imphal']
```

```python
temp=[]
city=[]
weather=[]
k=0
for i in caps:
    z=requests.get("http://api.openweathermap.org/data/2.5/weather?q="+i+"&APPID=bafdecd14e244eeadeabf917b9a7b8cf")
    dic=json.loads(z.content.decode())
    print(type(z.content.decode()))
    if dic["cod"]==200:
        k+=1
        print(k)
        print("city name:",dic["name"])
        print("temperature:",dic["main"]["temp"])
        print("weather:",dic["weather"][0]["main"])
        temp.append(dic["main"]["temp"])
        city.append(dic["name"])
        weather.append(dic["weather"][0]["main"])
    else:
        print(i,"spelling is incorrect")
```

```python
import pandas as pd
dt=pd.DataFrame({"city":city,"temp":temp,"weather":weather})
dt
```

```python
dt.weather.unique()
wea=[]
for w in dt.weather.unique():
    wea.append(dt[dt.weather==w])
```
