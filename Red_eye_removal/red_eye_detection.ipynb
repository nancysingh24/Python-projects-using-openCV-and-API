```python
import requests
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
import cv2 as cv
import io
```

```python
ori=cv.imread("red1.jpg")
plt.imshow(ori)
```

```python
img_rgb=cv.cvtColor(ori,cv.COLOR_BGR2RGB)
plt.imshow(img_rgb)
img_gray=cv.cvtColor(img_rgb,cv.COLOR_RGB2GRAY)
plt.imshow(img_gray,cmap="gray")
```

```python
detect=cv.CascadeClassifier("eye_haar.xml")
```

```python
rs=detect.detectMultiScale(img_gray,scaleFactor=1.4,minNeighbors=10,minSize=(50,50))
```

```python
rs.shape
```

```python
pq=img_rgb.copy()                                         
for i in rs:
    x,y,w,h=i
    cv.rectangle(pq,(x,y),(x+w,y+h),(255,0,0),5)
    plt.imshow(pq)
```
