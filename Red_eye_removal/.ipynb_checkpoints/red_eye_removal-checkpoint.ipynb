```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
import requests
import io

red1=cv.imread("red1.jpg")
plt.imshow(red1)




```

```python
img_rgb=cv.cvtColor(red1,cv.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
image_array=np.array(img_rgb)
```

```python
gray=cv.cvtColor(img_rgb,cv.COLOR_RGB2GRAY)
plt.imshow(gray,cmap="gray")
```

```python
detector=cv.CascadeClassifier("eye_haar.xml")
```

```python
eyes=detector.detectMultiScale(gray,scaleFactor=1.4,minNeighbors=10,minSize=(50,50))
```

```python
eyes.shape
```

```python
ey_copy = img_rgb.copy()
for ey in eyes:
    x,y,w,h=ey
    cv.rectangle(ey_copy,(x,y),(x+w,y+h),(255,0,0),5)
    plt.imshow(ey_copy)
```

```python
#crop eyes
eyes_a=[]
for eye in eyes:
    x,y,w,h=eye
    ey=img_rgb[y:y+h,x:x+w]
    eyes_a.append(ey)
    
f,a=plt.subplots(1,2)
a[0].imshow(eyes_a[0])
a[1].imshow(eyes_a[1])
```

```python
#spliting in the rgb format
#split eye image into 3 channels
eye1=eyes_a[0]
r=eye1[:,:,0]
g=eye1[:,:,1]
b=eye1[:,:,2]
f,a=plt.subplots(1,3,figsize=(12,5))
a[0].imshow(r,cmap="gray")
a[0].set_title("red")
a[1].imshow(g,cmap="gray")
a[1].set_title("green")
a[2].imshow(b,cmap="gray")
a[2].set_title("blue")
```

```python
#thresholding
red_thres=(r>130)
blue_green_thres=(r>cv.add(b,g))
combined_thres=red_thres & blue_green_thres
  
f,a=plt.subplots(1,3,figsize=(10,5))
a[0].imshow(red_thres,cmap="gray")
a[0].set_title("red thresholding")
a[1].imshow(blue_green_thres,cmap="gray")
a[1].set_title("blue_green thresholding")
a[2].imshow(combined_thres,cmap="gray")
a[2].set_title("combined thresholding")
```

```python
#filling center dot
def fill_center_dot(mask):
    #making copy of mask in which we are going to fill
    flood=mask.copy()
    h,w=flood.shape[:2]
    #creating temp mask for the flood fill function
    maskTemp=np.zeros((h+2,w+2),np.uint8)
    #floodfill algo
    cv.floodFill(flood,maskTemp,(0,0),255)
    #changing white pixel to black and vice versa
    inverse_flood=cv.bitwise_not(flood)
    #bitwise_string
    return inverse_flood | mask
```

```python
mask=fill_center_dot(combined_thres.astype("uint8"))
mask=cv.dilate(mask,None,iterations=3)
plt.imshow(mask,cmap="gray")
```

```python
#final output
mean_bg=cv.add(b,g)/2
mask=mask.astype("bool")
co=r.copy()
np.copyto(co,mean_bg.astype("uint8"),where=mask)
#ccombining three layers
final=np.dstack((co,g,b))
plt.imshow(final,cmap="gray")
```
